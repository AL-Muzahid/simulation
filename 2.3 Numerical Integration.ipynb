```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return x**3
```

```python
dotx = [] #empty list for x axis
doty = [] #empty list for y axis
n=0
m=0
iterations = int(input()) # 50000
for i in range(iterations):
    rx = random.randint(200,500)
    rx = rx/100
    ry = random.randint(0,150)
    dotx.append(rx)
    doty.append(ry)
    if ry<=func(rx):
        m+=1
    n+=1
```

```python
area = (m/n)*(3*150)
area
```

```python
x = []
y = []
i=0
h=0.2
while i<10:
    val = i
    x.append(val)
    y.append(func(val))
    i += h
    i = round(i,2)
plt.scatter(dotx,doty)
plt.plot(x,y)
plt.show()
```
